```python
# reviews_level plot and lectures_level plot

def plotdata(data,feature):
    plt.figure(figsize=(10,7))
    plt.title("Plot of {} per level per subject".format(feature))
    ax = sns.barplot(x = 'level',y = feature,data=data,hue = 'subject')
    for label in ax.containers:
        ax.bar_label(label)
    plt.show()
    
    
featureslist = ['num_subscribers', 'num_reviews', 'num_lectures']
for feature in featureslist:
    plotdata(data,feature)
```

```python
# we need to extract year,month,date from the published timestamp

# so for that i will extract the year-month-date from the timestamp 

data['published_date'] = data['published_timestamp'].apply(lambda x:x.split('T')[0])
data.head(3)
```

```python
data[data['published_date']=='3 hours']
```

```python
# dropping of that

data = data.drop(data.index[2066])

# checking if the row has been dropped or not

data[data['published_date']=='3 hours']
```

```python
# converting the published date to pandas datetime object

data['published_date'] = pd.to_datetime(data['published_date'],format="%Y-%m-%d")
```

```python
data['Year'] = data['published_date'].dt.year

data['Month'] = data['published_date'].dt.month

data['Day'] = data['published_date'].dt.day

data['Month_name'] = data['published_date'].dt.month_name()


data.head()
```

```python
# profit year wise

plt.figure(figsize=(15,7))
data.groupby(['Year'])['profit'].sum().plot(kind = 'bar')
plt.xticks(fontsize = 20,fontweight = 'bold',rotation = 45)
plt.yticks(fontsize = 20,fontweight = 'bold',rotation = 45)
plt.xlabel('Year',fontsize = 20,fontweight = 'bold',
           color = 'blue')
plt.ylabel('Profit',fontsize = 20,fontweight = 'bold',
           color = 'blue')
```

```python
# profit month wise

plt.figure(figsize=(15,7))
data.groupby(['Month_name'])['profit'].sum().plot(kind = 'bar')
plt.xticks(fontsize = 20,fontweight = 'bold',rotation = 45)
plt.yticks(fontsize = 20,fontweight = 'bold',rotation = 45)
plt.xlabel('Year',fontsize = 20,fontweight = 'bold',
           color = 'blue')
plt.ylabel('Profit',fontsize = 20,fontweight = 'bold',
           color = 'blue')
```
